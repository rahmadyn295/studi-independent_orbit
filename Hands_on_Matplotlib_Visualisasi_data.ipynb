```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#mempersiapkan data
url = "https://people.sc.fsu.edu/~jburkardt/data/csv/freshman_kgs.csv"
df = pd.read_csv(url)
df.head()
```

```python
#menghitung jumlah data pria
jumlah_pria = df['Sex'].value_counts()['M']

##menghitung jumlah data pria
jumlah_wanita = df['Sex'].value_counts()['F']

jumlah_per_gender = [jumlah_pria,jumlah_wanita]
gender = ['Pria','Wanita']

#Membuat Bar Chart
plt.bar(gender,jumlah_per_gender)
plt.show()
```

```python
#menghitung jumlah data pria
jumlah_pria = df['Sex'].value_counts()['M']

##menghitung jumlah data pria
jumlah_wanita = df['Sex'].value_counts()['F']

jumlah_per_gender = [jumlah_pria,jumlah_wanita]
gender = ['Pria','Wanita']

#Membuat Bar Chart
plt.figure(figsize=(4,4))
plt.bar(gender,jumlah_per_gender, color=['blue','orange'])#jika ingin bar c hart
plt.title('Perbandingan Jumlah Wanita dan Pria')
plt.xlabel('jenis kelamin')
plt.ylabel('jumlah')
plt.show()

```

```python
#mengambil data Histogram
weight_sep = df[' "Weight (Sep)"']

#membuat histogram
plt.figure(figsize=(4,4))
plt.hist(weight_sep, color='green')#histogram
plt.title('Histogram Berat Badan (Sep)')
plt.xlabel('Frekuensi')
plt.ylabel('Berat')
plt.show()
```

```python
#mengambil data diagram garis
bmi_sep = df[' "BMI (Sep)"']

#membuat line chart
plt.figure(figsize=(8,3))
plt.plot(bmi_sep, color='red')#diagram garis
plt.title('BMI bulan September')
plt.xlabel('Peserta')
plt.ylabel('BMI')
plt.show()
```

```python
#mengambil 2 data diagram garis
bmi_sep = df[' "BMI (Sep)"']
bmi_apr = df[' "BMI (Apr)"']

#membuat line chart
plt.figure(figsize=(8,3))
plt.plot(bmi_sep, color='blue', label='BMI September')
plt.plot(bmi_apr, color='red', label='BMI April')#label disini digunakan untuk membuat legenda padadiagram garis
plt.title('BMI bulan September dan April')
plt.xlabel('Peserta')
plt.ylabel('BMI')
plt.legend()
plt.show()
```

```python
#mengambil data Scatter plot
berat_sep = df[' "Weight (Sep)"']
bmi_sep = df[' "BMI (Sep)"']

#membuat scatter plot
plt.figure(figsize=(6,4))
plt.scatter(berat_sep, bmi_sep)
plt.title('Hubungsan berat dan BMI bulan September')
plt.xlabel('Berat')
plt.ylabel('BMI')
plt.show()
```

```python
#mengambil data box plot
bmi_sep = df[' "BMI (Sep)"']

#membuat line chart
plt.figure(figsize=(4,3))
plt.boxplot(bmi_sep)
plt.title('Boxplot BMI bulan September')
plt.show()
```

```python
#menghitung jumlah data pria
jumlah_pria = df['Sex'].value_counts()['M']

##menghitung jumlah data pria
jumlah_wanita = df['Sex'].value_counts()['F']

jumlah_per_gender = [jumlah_pria,jumlah_wanita]
gender = ['Pria','Wanita']

#Membuat Pie Chart
plt.figure(figsize=(4,4))
plt.pie(jumlah_per_gender, labels= gender, colors=['lightblue','pink'], autopct = '%.2f')
plt.title('Perbandingan Jumlah Pria dan wanita')
plt.show()
```
